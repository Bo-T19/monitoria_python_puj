## Agregaciones

```python
precipitaciones=pd.read_excel('Precipitaciones_Totales_Mensuales_20250820.xlsx')

precipitaciones
```

```python
precipitaciones_filtrado=precipitaciones[['ESTACI√ìN','MUNICIPIO', 'A√ëO', 'ENERO','FEBRERO', 'MARZO', 'ABRIL', 'MAYO', 'JUNIO', 'JULIO', 'AGOSTO', 'SEPTIEMBRE', 'OCTUBRE', 'NOVIEMBRE', 'DICIEMBRE']]
```

```python
precipitaciones_agrupado=precipitaciones_filtrado.groupby(['MUNICIPIO','ESTACI√ìN']).mean()
```

## Guardar mi dataframe

```python
precipitaciones_agrupado.to_excel('PRECIPITACIONES.xlsx',sheet_name="DATA")

help(precipitaciones_agrupado.to_excel)
```

## Ejemplo transacciones

```python
transacciones = pd.read_excel('Transacciones_1000.xlsx')
topes=pd.read_excel('Topes.xlsx')
```

```python
topes
```

```python
transacciones['Mes']=transacciones['Fecha'].dt.month
transacciones_limpio=transacciones[['Nombre persona','Transacci√≥n','Mes']]

transacciones_agrupadas=transacciones_limpio.groupby(['Nombre persona','Mes']).sum()

df_final = transacciones_agrupadas.merge(topes, on='Nombre persona', how='left')

df_final['Pas√≥ del tope'] = df_final['Transacci√≥n'] > df_final['Tope transacciones']

df_final
```
