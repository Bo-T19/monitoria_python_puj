## Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(-10,10,100)
y1=np.log(x)
y2=np.cos(x)
```

```python
# Primera gráfica
fig, ax=plt.subplots(figsize=(10,5))

ax.plot(x,y1,label='Logaritmo', color="#8F00FD", linestyle=':', marker='x', )
ax.plot(x,y2, label='Coseo',color="#00FDBA" )

ax.set_title('Funciones matemáticas',fontsize=24)
ax.set_ylabel('Eje y')
ax.set_xlabel('Eje x')
ax.set_xlim(-10,10)
ax.set_ylim(-2,5)

ax.grid()


ax.legend()

plt.show()
```

## Seaborn

```python
import seaborn as sns
```

```python
train
```

### Boxplots (caja y bigotes)

```python
#Box plot
fig, ax = plt.subplots(figsize=(10,3))

sns.boxplot(data=train, x='edadloc', ax=ax )
```

```python
#Box plot
fig, ax = plt.subplots(figsize=(10,8))

ax.set_title('Boxplots de nómina de acuerdo con tamaño de mercado',fontsize=12)
ax.set_ylabel('Nomina (COP)')
ax.set_xlabel('Tamaño de mercado')

sns.boxplot(data=train, x='tamamer',y='nomina', ax=ax , color="red")
```

### Scatterplots

```python
#Box plot
fig, ax = plt.subplots(figsize=(10,8))

ax.set_title('Scatterplot',fontsize=24)
ax.set_ylabel('Nomina (COP)')
ax.set_xlabel('Edad del local (años)')

sns.scatterplot(data=train, x='edadloc',y='nomina',color='red', marker='^', ax=ax )
```

### Histogramas

```python
#Box plot
fig, ax = plt.subplots(figsize=(10,8))

ax.set_title('Histograma',fontsize=24)
ax.set_ylabel('Frecuencia')
ax.set_xlabel('Edad del local')

sns.histplot(data=train, x='edadloc', kde=True)
```

### Barras 

```python
#Box plot
fig, ax = plt.subplots(figsize=(10,8))

ax.set_title('Barras',fontsize=24)
ax.set_ylabel('Frecuencia')
ax.set_xlabel('Tamaño de mercado')

sns.countplot(data=train, x='tamamer')
```
