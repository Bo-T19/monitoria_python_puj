```python
# From scikit learn import (NAMESPACE) datasets
from sklearn import datasets
```

```python
cancer = datasets.load_breast_cancer()

#Crear data frame
X= pd.DataFrame(cancer.data, columns = cancer.feature_names)
X
```

```python
Y= pd.Series(cancer.target, name ='target')
Y
```

```python
#dataset de prueba
df=pd.read_excel('dataset_prueba.xlsx')
df_sexo=df.drop(['Edad'], axis = 1)
df_sexo
```

```python
one_hot= OneHotEncoder()
df_one_hot=one_hot.fit_transform(df_sexo)

print(df_one_hot)
```

```python
df_one_hot_pandas=pd.DataFrame.sparse.from_spmatrix(df_one_hot)
df=pd.concat([df,df_one_hot_pandas], axis = 1)
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size=0.2, random_state=42)

X_train
```

```python
#Entrenar modelo
log_reg= LogisticRegression(max_iter=5000)

log_reg.fit(X_train,y_train)
```

```python
y_predicted= log_reg.predict(X_test)
y_predicted
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy= accuracy_score(y_test,y_predicted)
print(classification_report(y_test,y_predicted))
```
