## Modelo lineal

```python
df_train.drop(['id'], axis =1, inplace=True)
```

```python
df_train
```

```python
y=df_train['MedHouseVal']
y
```

```python
x= df_train.drop(['MedHouseVal'], axis =1)
x
```

```python
regresion = LinearRegression()
```

```python
regresion.fit(x,y)
```

```python
regresion.coef_
```

```python
regresion.intercept_
```

## Test

```python
test=pd.read_csv('test.csv')
test
```

```python
test.drop(['id'], axis=1, inplace=True)
```

```python
test
```

```python
prediccion=regresion.predict(test)
```

```python
prediccion
```

```python
submission = pd.DataFrame({'id':test['id'], 'MedHouseVal':prediccion})
```

## Evaluación regresión

```python
import statsmodels.api as sm
```

```python
x=sm.add_constant(x)
```

```python
regresion_stats_models_1=sm.OLS(y,x)
results = regresion_stats_models_1.fit()
```

```python
r_cuadrado=results.rsquared
r_cuadrado
```

```python
r_cuadrado_ajustado=results.rsquared_adj
r_cuadrado_ajustado
```

```python
aic=results.aic
aic
```

```python
results.summary()
```
